##### Plot the list of vectors

```python
def plotting_vectors(vect):
    for i in vect:
        print(i[0], i[1], i[2], i[3])
#         plt.quiver(i[0], i[1], i[2], scale_units="xy", angles="xy", scale=1, color="red") doesnt work
# Because the coordinate axis are 4 points of 2D space but the above code has only 3 elements of a 2D space
        plt.quiver(i[0], i[1], i[2], i[3], scale_units="xy", angles="xy", scale=1, color="red")
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    plt.show()
        
vectorList = [(0,0,3,6), (0,0,-3,6), (0,0,-3, -6), (0,0,3,-6), (1,1,6,6)]
plotting_vectors(vectorList)
```

###### Projecting one vector on other visualized!

```python
def plot_vectors(vectors):
    colors = ['b', 'y', 'g']
    n = 0
    for i in vectors:
        plt.quiver(i[0], i[1], i[2], i[3], scale_units="xy", angles="xy", scale=1, color=colors[n])
        n += 1
    plt.xlim(-1,10)
    plt.ylim(-1,10)
    plt.show()
```

```python
plot_vectors([(0,0,7,3), (0,0,8,6)])
```

```python
dot_prod = np.dot((0,0,7,3), (0,0,8,6))
print(dot_prod)
```

```python
proj_mag = dot_prod / np.linalg.norm((0,0,8,6))
print(proj_mag)
```

```python
proj_coord = (proj_mag/np.linalg.norm((0,0,8,6)))*np.asarray([0,0,8,6])
print(proj_coord)
```

```python
plot_vectors([(0,0,7,3), (0,0,8,6), (0,0,5.92, 4.44)])
```
