```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
```

```python
datasets = sklearn.datasets.load_breast_cancer()
```

```python
X = datasets.data
Y = datasets.target
```

## To print the shapes of the dataset:

```python
print(X.shape)
print(Y.shape)
```

###### 

## To Convert the array dataset to pandas Dataframe
#### Note: While converting to Dataframe, we must pass the column attribute information also
###### Note: describe and head are pandas functions so () are important!

```python
dataset = pd.DataFrame(datasets.data, columns=datasets.feature_names)
```

```python
dataset.describe()
```

###### 

###### 

## To add a new column to the datasets 
##### (Note while visualizing the dataset, shouldn't use dataset.head() inside the print statement)

```python
dataset['class'] = datasets.target
```

```python
dataset.head()
```

## To find the number of entries for each values

```python
print(dataset['class'].value_counts())
```

```python
print(datasets.target_names)
```

### 

## Splitting of dataset

```python
from sklearn.model_selection import train_test_split
```

```python
x = dataset.drop('class', axis=1)
y = dataset['class']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, stratify= y, random_state=1 , shuffle=True, test_size=0.2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
print(y.mean(), y_train.mean(), y_test.mean())
```

```python
print(y.shape, y_train.shape, y_test.shape)
```

### 

## Binarisation of Inputs

```python
plt.plot(dataset)
plt.show()
```

```python
plt.plot(dataset.T, '.')
plt.xticks(rotation="vertical")
plt.show()
```

#### To binarise a single column:

```python
x_binarised_3_train = x_train['mean area'].map(lambda x:0 if x < 1000 else 1)
```

```python
plt.plot(x_binarised_3_train, '*')
plt.show()
```

###### 

## To binarize all the columns in one shot using pandas

```python
x_train_bin = x_train.apply(pd.cut, bins=2 ,labels=[0,1])
x_test_bin = x_test.apply(pd.cut, bins=2 ,labels=[0,1])
```

```python
plt.plot(x_train_bin)
plt.show()
```

```python
plt.plot(x_train_bin, '.')
plt.show()
```

```python
x_train.describe()
```

```python
x_train_bin.describe()
```

```python
x_train_bin['mean radius'].value_counts()
```

```python
plt.plot(x_train_bin.T, '.')
plt.xticks(rotation='vertical')
plt.show()
```

## To convert the Dataframe type back to numpy arrays:

```python
x_train = x_train_bin.values
x_test = x_test_bin.values
```
