```python
import cv2
```

```python
cap = cv2.VideoCapture(0)
```

```python
while True:
    photo = cv2.imread('face_swapping.jpg')
    face_model = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
    face_train = face_model.detectMultiScale(photo)
    if(len(face_train)) == 0:
        pass
    else:
        x1 = face_train[0][0]
        y1 = face_train[0][1]
        x2 = x1+face_train[0][2]
        y2 = y1+face_train[0][3]
        
        x3 = face_train[1][0]
        y3 = face_train[1][1]
        x4 = x3+face_train[1][2]
        y4 = y3+face_train[1][3]
        
        x5 = face_train[2][0]
        y5 = face_train[2][1]
        x6 = x5+face_train[2][2]
        y6 = y5+face_train[2][3]
        
        cv2.rectangle(photo, (x1,y1), (x2,y2), [255,0,0], 3)
        cv2.rectangle(photo, (x3,y3), (x4,y4), [0,255,0], 3)
        cv2.rectangle(photo, (x5,y5), (x6,y6), [0,0,255], 3)
        cv2.imshow('hi', photo)
        if cv2.waitKey(1) == 13:
            break
cv2.destroyAllWindows()
```

```python
cap.release()
```
