```python
import pandas as pd
```

```python
data =  pd.read_csv('Churn_Modelling.csv')
```

```python
data.head(15)
```

```python
dataset = data[['CreditScore', 'Geography', 'Gender', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard', 'IsActiveMember', 'EstimatedSalary']]
```

```python
dataset
```

```python
geo = pd.get_dummies(dataset['Geography'], drop_first=True)
```

```python
geo.columns
```

```python
x = dataset.iloc[ : , [0,3,4,5,6,7,8]]
```

```python
x
```

```python
x = x.iloc[:, 0:9]
```

```python
x[['Germany', 'Spain']] = geo
```

```python
x
```

```python
gender = pd.get_dummies(dataset['Gender'], drop_first=True)
```

```python
gender
```

```python
x = x.iloc[: , 0:10]
```

```python
x[['gender']] = gender
```

```python
x
```

```python
y = dataset['EstimatedSalary']
```

```python
y
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split( x, y, test_size=0.15, random_state=42)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_test.shape
```

```python
y_pred.shape
```

```python
error = mean_squared_error(y_test, y_pred)
```
