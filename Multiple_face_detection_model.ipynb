```python
import cv2
```

```python
cap = cv2.VideoCapture(0)
face_model = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
#photo =  cv2.imread('face_swapping.jpg')
```

```python
while True:
    status, photo = cap.read()
    face_train = face_model.detectMultiScale(photo)
    for i in range(0,len(face_train)):
            for j in range(4):
                if len(face_train) ==0:
                    pass
                else:
                    x1= face_train[i][0]
                    y1 = face_train[i][1]
                    x2 = x1 + face_train[i][2]
                    y2 = y1 + face_train[i][3]
                    photo = cv2.rectangle(photo, (x1,y1), (x2,y2), [0,255,0], 3)
                    cv2.imshow('Attention please', photo)
    if cv2.waitKey(1) == 13:
            break
cv2.destroyAllWindows()
        
```

```python
cap.release()
```
